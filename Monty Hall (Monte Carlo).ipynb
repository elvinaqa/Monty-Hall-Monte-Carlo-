```
import random
```

```
def choose_door(num_doors):
    # Here's where we get user input. 
  print("There are " + str(num_doors) + " doors. Behind one door is a prize. The other " + str(num_doors - 1) + " doors are empty.")
  print("Pick one door.")
  pick = int(input("> "))
  return pick
```

```
def is_prize(prize, pick):
  if prize == pick:
    return True
  return False
```

```
def is_fake(n_doors, pick):
  fake_door = random.randint(0, n_doors)
# keep opening other doors randomly
  if fake_door == pick:
    is_fake(n_doors, pick)
  return fake_door
```

```
def change_door(pick, second_pick):
  print("U chose Door" + str(pick))
  print("All other doors are open, apart " + str(second_pick))
  print("Wanna change choice?")
  change = input("y/n: ")
  return change
```

```
def success_or_fail(second_pick, prize, pick, change):
  count = 0
  if change == 'y':
    print("Switched to Door ", second_pick)
  elif change == 'n':
    print("You ignore to change Door", pick)

  print("Prized door is", prize)

  if is_prize(prize, second_pick):
    count+=1
    print(count, "Times Success")
  else:
    print("Fail")
```

```
def __main__():
  
  n_doors = 3

  prize = random.randint(0, n_doors-1)

  pick = choose_door(3)

  picked_prize = is_prize(prize, pick)

  if picked_prize:

    fake_door = is_fake(n_doors, pick)

    change = change_door(pick, fake_door)

    success_or_fail(fake_door, prize, pick, change)
  elif not picked_prize:

    change = change_door(pick, prize)

    success_or_fail(prize, prize, pick, change)

if __name__ == "__main__":
  __main__()
```

```
y = int(input("How many times to run?> "))
x = 0
win = 0

while x<y:
  win = __main__()
  x = x+1
print("WINS: ", str(win))
print("Percentage by changing: ", int(win/y)*100)
```

```
def all():
  win = 0
  n_doors = 3

  prize = random.randint(0, n_doors-1)

  pick = choose_door(3)

  picked_prize = is_prize(prize, pick)

  if picked_prize:

    fake_door = is_fake(n_doors, pick)

    change = change_door(pick, fake_door)

    success_or_fail(fake_door, prize, pick, change)
    return win
  elif not picked_prize:

    change = change_door(pick, prize)

    success_or_fail(prize, prize, pick, change)
    win+1
y = int(input("How many times do you want to run the simulation?> "))
x = 0
win = 0
while x < y:
    win = all()
    x = x+1
print("WINS: " , win)
print("Percent Won by Switching: ", win, "out of", y, " number of times","%")
```
